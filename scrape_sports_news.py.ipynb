```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
```

```python
date = input('Enter date MM/DD/YY: ')
URL = f'https://www.yallakora.com/match-center/%D9%85%D8%B1%D9%83%D8%B2-%D8%A7%D9%84%D9%85%D8%A8%D8%A7%D8%B1%D9%8A%D8%A7%D8%AA?date={date}#'
webpage = requests.get(URL)
webpage
```

```python
def Scrap(webpage):

  src = webpage.content
  soup = BeautifulSoup(src, 'lxml')
  matches_details = []

  championships = soup.find_all('div', {'class': 'matchCard'})

  for i in range(len(championships)):
    championship_title = championships[i].find('h2').text.strip()

    for ul_element in championships[i].find_all('div', {'class':'ul'}):
      matches_list = ul_element.find_all('div', {'class':'item finish liItem'}) + ul_element.find_all('div', {'class':'item now liItem'}) + ul_element.find_all('div', {'class':'item future liItem'})
      for match in matches_list:
        team_a = match.find('div', {'class':'teamA'}).text.strip()
        team_b = match.find('div', {'class':'teamB'}).text.strip()
        score = match.find_all('span', {'class':'score'})
        score_a = score[0].text.strip()
        score_b = score[1].text.strip()
        time = match.find('span', {'class':'time'}).text.strip()
        matches_details.append({
            'championship_title': championship_title,
            'Teams': f'{team_a} vs {team_b}',
            'Score': f'{score_a} - {score_b}',
            'match_time': time
        })

  df = pd.DataFrame(matches_details)
  df.to_csv('matches.csv')
```

```python
Scrap(webpage)
```
